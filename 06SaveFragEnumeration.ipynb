```python
enumeration_data=[]
for idx,a_ms_data in ms_pcp_interp.iterrows():
    enume_smiles={}
    enume_smiles["iupac_name"]=a_ms_data["iupac_name"]
    
    mol=Chem.MolFromSmiles(a_ms_data["canonical_smiles"])
    mol=Chem.AddHs(mol)
    
    print("iupac_name:",a_ms_data["iupac_name"])
    print("--- start enumeration ---")
    CP_t=fragenum.CombProducts(b_exist_atm=True,b_exist_rad=True)
    CP_t.set_mol(mol)
    CP_t.GenProComb()
    print("--- end enumeration ---")
    print()
    
    enumurated_smiles_list=[]
    for j in range(len(CP_t.l_prod)):
        enumurated_smiles_list.append(CP_t.l_prod[j].smiles_list)
    enume_smiles["enumurated_smiles_list"]=enumurated_smiles_list
    enumeration_data.append(enume_smiles)
```
